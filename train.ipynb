```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
```

```python
X, y = make_regression(n_samples=900, n_features=1, noise=5, random_state=4)
```

```python
plt.scatter(X, y, color='blue')
plt.show
```

```python
import numpy as np
```

```python
data = np.concatenate((X, y.reshape(-1, 1)), axis=1)

np.savetxt("dataset_training.csv", data, delimiter=",", header="feature,traget", comments='')
```

```python
train_X, test_X, train_y, test_y = train_test_split(X, y, train_size=0.85, random_state=3)
```

```python
model = LinearRegression()
pred = model.fit(train_X, train_y).predict(test_X)
```

```python
test_y
```

```python
from sklearn.metrics import mean_absolute_percentage_error

mape = mean_absolute_percentage_error(test_y, pred)

mape
```

```python
import joblib

joblib.dump(model, 'models/model.pkl')

print("Saved!")
```

```python
model = joblib.load("models/model.pkl")

predictions = model.predict(test_X)

mape = mean_absolute_percentage_error(test_y, predictions)

mape
```
